```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = yf.download("AAPL", start="2025-01-01", end="2026-01-01")

data.head()
```

```python
data["Close"].plot(figsize=(12,5))
plt.title("Apple Share Price")
plt.show()
```
